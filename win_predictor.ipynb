```python
team=['Sunrisers Hyderabad',
      'Mumbai Indians',
      'Royal Challengers Bangalore',
      'Kolkata Knight Riders',
      'Kings XI Punjab',
      'Chennai Super Kings',
      'Rajasthan Royals',
      'Delhi Capitals']
```

```python
match_df["team1"]=match_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match_df["team1"]=match_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")
match_df["team1"]=match_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match_df["team1"]=match_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
```

```python
match_df=match_df[match_df["team1"].isin(team)]
match_df=match_df[match_df["team2"].isin(team)]
```

```python
match_df.shape
```

```python
match_df["dl_applied"].value_counts()
```

```python
match_df=match_df[match_df["dl_applied"]==0]
```

```python
match_df=match_df[["match_id","city","winner","total_runs"]]
```

```python
match_df
```

```python
delivery_df=match_df.merge(delivery,on="match_id")
```

```python
delivery_df[delivery_df['inning']==2].head()
```

```python
delivery_df=delivery_df[delivery_df["inning"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df.info()
```

```python
delivery_df.isnull().sum()
```

```python
delivery_df["currect_score"]=delivery_df.groupby("match_id")["total_runs_y"].cumsum()
```

```python
delivery_df[delivery_df["match_id"]==1].head(60)
```

```python
delivery_df["run_left"]=delivery_df["total_runs_x"]-delivery_df["currect_score"]
```

```python
delivery_df["ball_left"]=120-((delivery_df["over"]-1)*6+delivery_df["ball"])
```

```python
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].fillna("0")
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].apply(lambda x:0 if x=="0" else 1)
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].astype(int)
```

```python
delivery_df["wicket"]=delivery_df.groupby("match_id")["player_dismissed"].cumsum().values
```

```python
delivery_df["player_dismissed"].value_counts()
```

```python
delivery_df["remaing_wicket"]=10-delivery_df["wicket"]
```

```python
delivery_df["crr"]=(delivery_df["currect_score"]/(120-delivery_df["ball_left"]))*6
```

```python
delivery_df["rrr"]=(delivery_df["run_left"]/(delivery_df["ball_left"]))*6
```

```python
delivery_df
```

```python
def result(row):
   return 1 if row["batting_team"]==row["winner"] else 0
```

```python
delivery_df["result"]=delivery_df.apply(result,axis=1)
```

```python
delivery_df.head(1)
```

```python
delivery_df.isna().sum()
```

```python
delivery_df["city"].fillna(method="backfill",inplace=True)
```

```python
final_df=delivery_df[["batting_team","bowling_team","city","currect_score","run_left","ball_left","remaing_wicket","total_runs_x","crr",'rrr','result']]
```

```python
final_df.dropna(inplace=True)
final_df.isnull().sum()
#final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample(3)
```

```python
final_df.shape
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df["ball_left"]!=0]
```

```python
final_df.shape
```

```python
x=final_df.iloc[:,:-1]
y=final_df["result"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
step2=LogisticRegression(solver="liblinear")
```

```python
step1=ColumnTransformer([('trf',OneHotEncoder(sparse_output=False,drop="first"),["batting_team","bowling_team","city"])],remainder="passthrough")
```

```python
pipe=Pipeline([("step1",step1),
             ("step2",step2)])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl', 'wb'))
```
